```python
import pandas as pd
```

```python
# Carga el dataset de trabajo
df_merged = pd.read_csv("datos/df_merged.csv")
df_merged.drop(columns=['Unnamed: 0'], inplace=True)
df_merged.head()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_movie = df_merged['type'] == 'Movie'
condicion_2021 = df_merged['release_year'] == 2021

df_1 = df_merged[(condicion_movie) & (condicion_2021)]

df_1.head()
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
condicion_2 = df_merged['duration'] == '90 min'
df_2 = df_merged[condicion_2][['title', 'duration']]
df_2.head()
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
condicion_india = df_merged['country'] == 'India'
condicion_despues_2018 = df_merged['release_year'] > 2018

df_3 = df_merged[(condicion_india) & (condicion_despues_2018)]
df_3.head()
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_4 = df_merged['rating'].isin(['TV-14', 'R'])
df_4 = df_merged[condicion_4]
df_4.head()
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
condicion_5 = df_merged['director'].isin(['Christopher Nolan', 'Martin Scorsese'])

df_5 = df_merged[condicion_5]

df_5.head()
```

```python
df_5.groupby('director')['title'].count().reset_index()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
cond_united = df_merged['country'].str.contains('United States', case=False)
cond_kingdom = df_merged['country'].str.contains('United Kingdom', case=False)
cond_canada = df_merged['country'].str.contains('Canada', case=False)

df_6 = df_merged[(cond_united) | (cond_kingdom) | (cond_canada)]
df_6.head()
```

```python
print("De estas, tienen varios lenguages\n")
df_6[df_6['Language'].str.contains('/', na=False)].shape[0]
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
condicion_7 = df_merged['release_year'].between(2015, 2020, inclusive='both')
df_7 = df_merged[condicion_7]
df_7.head()
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
condicion_8 = df_merged['IMDB Score'].between(6.5, 8.0)
condicion_nulo = df_merged['duration'].isnull()

df_8 = df_merged[(condicion_8) & (condicion_nulo)]
df_8.head()
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
condicion_9 = df_merged['duration'].isin(['90 min', '150 min'])

df_9 = df_merged[condicion_9]
df_9.sample(5)
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
condicion_10 = df_merged['title'].str.contains(pat =r"\b^[Ll]ove", case=False, na=False)
condicion_tv_show = df_merged['type'] == 'TV Show'

df_10 = df_merged[(condicion_10) & (condicion_tv_show)]
df_10.head()
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
condicion_11 = df_merged['director'].str.contains(' Johnson', case=False, na=False)

df_11 = df_merged[condicion_11]
df_11.head()
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

### Este lo ha sacado buscando en internet como se llama el documental de Myke Tyson

```python
condicion_12 = df_merged['listed_in'].str.contains('Documentaries', case=False, na=False)
condicion_tyson = df_merged['title'].str.contains('tyson', case=False, na=False)

df_12 = df_merged[(condicion_12) & (condicion_tyson)]
df_12
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
cond_title = df_merged['title'].notnull() == True
cond_country = df_merged['country'].notnull() == True
cond_IMBD = df_merged['IMDB Score'].notnull() == True

df_20 = df_merged[cond_title & cond_country & cond_IMBD]
df_20[['title', 'country', 'IMDB Score']]
```
